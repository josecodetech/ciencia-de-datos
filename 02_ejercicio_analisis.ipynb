```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/ventas_mensuales.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Mes'],df['Ventas'],marker='o',color='b',linestyle='-',linewidth=2)
plt.title('Ventas mensuales')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
media_ventas = df['Ventas'].mean()
print(f'Media de las ventas = {media_ventas:.2f}')
```

```python
mes_mas_ventas = df[df['Ventas']==df['Ventas'].max()]['Mes'].values[0]
print(f'Mes con mayores ventas = {mes_mas_ventas}')
```

```python
plt.figure(figsize=(10,6))
plt.bar(df['Mes'],df['Ventas'],color='g')
plt.title('Ventas mensuales')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.show()
```
