```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'A': [1,2,np.nan,4],
    'B': [5,np.nan, np.nan, 8],
    'C': [10,11,12,13]
})
```

```python
df.isnull()
```

```python
# Borra filas con nulos
df_sin_nulos = df.dropna()
df_sin_nulos
```

```python
# Borrar columnas con nulos
df_sin_nulos = df.dropna(axis = 1)
df_sin_nulos
```

```python
# rellenar valores nulos
df_relleno = df.fillna(value = 0)
df_relleno
```

```python
# rellena con siguiente valor
df_relleno = df.fillna(method='ffill')
df_relleno
```

```python
# rellena con valor siguiente
df_relleno = df.fillna(method='bfill')
df_relleno
```

```python
# rellena con la media
df_relleno = df.fillna(df.mean())
df_relleno
```

```python
# rellena mediante interpolacion
df_relleno = df.interpolate()
df_relleno
```
